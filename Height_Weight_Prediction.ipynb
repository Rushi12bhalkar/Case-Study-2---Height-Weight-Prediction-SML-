```python

# supervised ML case study 2 : predicting Weight from Height

import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# output Result

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

# machine learning library
# functions for splitting dataset and cross validation scores on training set
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score  # find prediciton on training set
from sklearn.model_selection import StratifiedKFold   # create no of folds or splits

# functions to generate the output 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

# functions for algorithms implementation
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

ds = pd.read_csv(r'height-weight.csv')
print(ds)

# Step 1 :- understand and analyze the data
print(ds.shape)
print(ds.head(10))
print(ds.tail(10))
print(ds.describe())
print(ds.isnull().sum()) # find missing values in datasets
# datasets does not have any missing values

```

```python
# Step 2 :- visualize the data by using graphs.

plt.subplot(1,2,1)
x1 = ds.iloc[:, 0].values # gender 
y1 = ds.iloc[:, 2].values # weight
plt.grid(True)
plt.scatter(x1,y1,label='Gender',color='Green')
plt.xlabel('Gender')
plt.ylabel('Weight')
plt.title('Gender vs Weight')
plt.legend()
#plt.show()

plt.subplot(1,2,2)
x2 = ds.iloc[:, 0].values # gender
y2 = ds.iloc[:, 1].values # height
plt.grid(True)
plt.scatter(x2,y2,label='Gender',color='Red')
plt.xlabel('Gender')
plt.ylabel('Height')
plt.title('Gender vs Height')
plt.legend()
plt.show()


x3 = ds.iloc[:, 1].values # Height
y3 = ds.iloc[:, 2].values # Weight
plt.grid(True)
plt.scatter(x2,y2,label='Gender',color='MAGENTA')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs Height')
plt.legend()
plt.show()

import seaborn as sb
  
sb.lmplot(x = "Height", y = "Weight",  ci = None, data = ds)
plt.show()

```

```python
# Step 3 :- Data Transformation Step
# Convert Catagerical data into nummeric data [Gender-male/female]

ds['Gender'].replace('Female', 0, inplace=True)
ds['Gender'].replace('Male', 1, inplace=True)

print(ds.head())
print(ds.tail())

X = ds.iloc[:, :-1].values  # X - gender and height - input var
y = ds.iloc[:, 2].values # weight - output
```

```python
# Step 4 :- Splitting data into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# X_train - gender/ht - 80 %  - 8000
# X_test  - gender/ht - 20 %  - 2000
# y_train - weight    - 80 %  - 8000
# y_test  - weight    - 20 %  - 2000

# Step 5:- Trian the Regression Model
model =  KNeighborsClassifier()
model.fit(X_train, y_train) # train the model on the training set

# Step 6:- Make Prediction using test set
prediction = model.predict(X_test)
print(prediction)


# Explained variance score: 1 is perfect prediction
print('Accuracy score: {:.2f}'.format(model.score(X_test, y_test)*100)) 
print('Accuracy score: {:.2f} %'.format(model.score(X_test, y_test)*100))


# Step 7 :-  Predict my weight
myweight = model.predict([[0,185]]) # 104
print('My predicted weight = ',myweight) 



```
